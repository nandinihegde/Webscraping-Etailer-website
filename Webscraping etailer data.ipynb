# Web scraping from a etailer website

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
import pandas as pd
import os

```

```python
base_url = "https://www.forever21.com/us/shop/catalog/category/f21/dress"

# create a new Firefox session
driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.get(base_url)

```

```python

# To extract the first and last page numbers
# get internal HTML generated by Javascript
innerHTML = driver.execute_script("return document.body.innerHTML") 
#Transfer HTML to beautifulsoup
page_soup= BeautifulSoup(innerHTML,"html.parser")

paging = page_soup.find("div",{"id":"bottom-pager"}).find_all("li")
start_page = paging[1].text
last_page = paging[len(paging)-2].text
```
